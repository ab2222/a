```python
import numpy as np

X = np.array(([2, 9], [1, 5], [3, 6]), dtype=float)
y = np.array(([92], [86], [89]), dtype=float)
X = X/np.amax(X,axis=0) 
y = y/100

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivatives(x):
    return x * (1 - x)

epoch=5 
lr=0.1 

inputlayer_neurons = 2 
hiddenlayer_neurons = 3 
outputlayer_neurons = 1 

wh=np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
bh=np.random.uniform(size=(1,hiddenlayer_neurons))
wout=np.random.uniform(size=(hiddenlayer_neurons,outputlayer_neurons))
bout=np.random.uniform(size=(1,outputlayer_neurons))

for i in range(epoch):
    hinp=np.dot(X,wh)+ bh
    hlayer = sigmoid(hinp)
    outinp=np.dot(hlayer,wout)+bout
    output = sigmoid(outinp)
    
    EO = y-output
    outgrad = sigmoid_derivatives(output)
    d_output = EO * outgrad
    EH = d_output.dot(wout.T)
    hiddengrad = sigmoid_derivatives(hlayer)
    d_hiddenlayer = EH * hiddengrad
    
    wout += hlayer.T.dot(d_output) *lr   
    wh += X.T.dot(d_hiddenlayer) *lr
    
    print ("-----------Epoch", i+1, "Starts----------")
    print("Input: \n" + str(X)) 
    print("Actual Output: \n" + str(y))
    print("Predicted Output: \n" ,output)
    print ("-----------Epoch", i+1, "Ends----------\n")
        
print("Input: \n" + str(X)) 
print("Actual Output: \n" + str(y))
print("Predicted Output: \n" ,output)
```
